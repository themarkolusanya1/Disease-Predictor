```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
# 1. Load the dataset
file_path = "symptoms_Africa20.csv"
```

```python
# 2. Read the CSV file
df = pd.read_csv(file_path)
```

```python
# 3. Print dataset initialization details
print(f"Dataset Loaded Successfully: {file_path}")
print(f"Total Rows: {df.shape[0]} | Total Columns: {df.shape[1]}")
print(f"\nDetected Columns:\n{list(df.columns)}")
print("\nFirst 3 Rows of Raw Data:")
print(df.head(3))
```

```python
# 4. Separate Features (symptoms) and Target (disease)
X = df.drop(columns=["Disease"])
y = df["Disease"]
```

```python
# 5. Encode Categorical Symptom Features into Numeric (0/1)
X_encoded = pd.get_dummies(X, drop_first=True).astype(int)
```

```python
# 6. Encode Target Labels (Disease Names -> Numeric IDs)
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
# 7. Check total duplicate rows across the entire dataset
duplicate_count = df.duplicated().sum()
print(f"Total Exact Duplicate Rows Found: {duplicate_count}")
```

```python
# 8. Train / Test Split using shuffle=True (Preserves 100% of data & rare classes)
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, 
    y_encoded, 
    test_size=0.20, 
    random_state=42, 
    shuffle=True
)
```

```python
# 9. Verify Results
print(f"Cleaned Rows Retained: {len(X_encoded)} / {len(df)}")
print(f"X_train Shape        : {X_train.shape}")
print(f"X_test Shape         : {X_test.shape}")
print(f"y_train Shape        : {y_train.shape}")
print(f"y_test Shape         : {y_test.shape}")
```
